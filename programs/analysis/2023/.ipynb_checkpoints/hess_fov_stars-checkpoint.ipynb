# Star angular distribution to HESS targets

```python
from astropy.table import Table, vstack, unique
from astropy.coordinates import SkyCoord
import astropy.units as u

import matplotlib.pyplot as plt

import os

import numpy as np
```

## HESS targets

```python
crab_pos = SkyCoord(184.5574, -5.7845, unit=u.deg, frame='galactic')
pks2155_pos = SkyCoord(17.7373, -52.2477, unit=u.deg, frame='galactic')
gal_center = SkyCoord(359.9449, -0.0439, unit=u.deg, frame='galactic')
```

## Star catalogues

```python
catalogs_files = []
for f in os.scandir('/home/pedroivo/SII'):
    if f.path.endswith('.csv'):
        catalogs_files.append(f.path)
```

```python
cat_tables = [Table.read(fname) for fname in catalogs_files]
```

```python
star_catalog = unique(vstack(cat_tables), silent=True)
```

## Fetch star positions from catalogues

```python
star_pos = SkyCoord(ra=star_catalog['RA'] * u.hourangle, dec=star_catalog['DEC'] * u.deg, frame='icrs')
```

## Plot HESS FoV with the stars and the targets

```python
%matplotlib inline
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111)

ax.scatter(crab_pos.icrs.ra, crab_pos.icrs.dec, color='orange')
ax.scatter(star_pos.ra, star_pos.dec, color='white')
ax.scatter(pks2155_pos.icrs.ra, pks2155_pos.icrs.dec, color='green')
ax.scatter(gal_center.icrs.ra, gal_center.icrs.dec, marker='x', color='red')
#ax.set_yticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
#ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
ax.grid(True)
ax.set_xlabel(r'RA/$\degree$', fontsize=15)
ax.set_ylabel(r'Dec/$\degree$', fontsize=15)
ax.set_facecolor('darkblue')

fig.savefig('skymap.pdf')
```

## Calculate angular distance between stars and sources

```python
dist_to_crab = crab_pos.separation(star_pos) - 90 * u.deg
dist_to_pks2155 = pks2155_pos.separation(star_pos) - 90 * u.deg
dist_to_gcenter = gal_center.separation(star_pos) - 90 * u.deg
```

```python
fig, ax = plt.subplots(figsize=(12,10))

ax.hist(dist_to_crab, bins=36, range=(-90, 90) * u.deg, label='Crab', )
ax.hist(dist_to_pks2155, bins=36, range=(-90, 90)  * u.deg, alpha=0.5, label='PKS 2155-304')
ax.hist(dist_to_gcenter, bins=36, range=(-90, 90) * u.deg, alpha=0.3,  label='Gal. center')

ax.set_xlabel(r'Angular distance/$\degree$', fontsize=15)

fig.legend()
fig.savefig('ang_dist.pdf')
```
